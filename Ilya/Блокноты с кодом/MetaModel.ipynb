```python
import numpy as np  # для работы с массивами и математическими функциями
import pandas as pd  # для обработки и анализа данных, особенно табличных данных
from sklearn.linear_model import LogisticRegression


import optuna  # для оптимизации гиперпараметров модели с помощью продвинутых методов поиска
import lightgbm as lgb  # для использования алгоритма градиентного бустинга LightGBM
from sklearn.model_selection import train_test_split  # для разделения данных на обучающую и тестовую выборки

from sklearn.metrics import roc_auc_score
import xgboost as xgb
import joblib
```

```python
df_hack_test = pd.read_csv('..\DataSet\Test_датасет_альфа')
df_hack_train = pd.read_csv('..\DataSet\Train_датасет_альфа')
```

```python
data_test = df_hack_test.iloc[:, 1:]
```

```python
df_hack_train
```

```python
X = df_hack_train.drop(['target','smpl'], axis=1)
y = df_hack_train['target']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)

# 2. Разделение тренировочной выборки на тренировочную и валидационную для метамодели
X_train_meta, X_valid_meta, y_train_meta, y_valid_meta = train_test_split(X_train, y_train, test_size=0.2, random_state=42)
```

```python
model1 = joblib.load("..\Предобученные модели\Test_Alpha_Hack_LightGBM_Optuna.pl")
model2 = joblib.load("..\Предобученные модели\Test_Alpha_Hack_LightGBM_Optuna1.pl")
```

```python
X = data_test.iloc[:len(data_test)//2, :]
X2 = data_test.iloc[len(data_test)//2:, :]
```

```python
X
```

```python
# Получение предсказаний для LightGBM моделей
preds_1 = model1.predict(X)
preds_2 = model2.predict(X2, predict_disable_shape_check=True)
```

```python
meta_features = np.column_stack([preds_1, preds_2])
```

```python
# Метамодель — логистическая регрессия# Создание мета-признаков
meta_features = np.column_stack([preds_1, preds_2])

# Функция для оптимизации гиперпараметров метамодели с помощью Optuna
def objective(trial):
    # Гиперпараметры логистической регрессии
    C = trial.suggest_float("C", 1e-4, 1e2, log=True)
    solver = trial.suggest_categorical("solver", ["lbfgs", "liblinear"])

    # Обучение метамодели
    meta_model = LogisticRegression(C=C, solver=solver, random_state=42, max_iter=1000)
    meta_model.fit(X_train_meta, y_train_meta)

    # Предсказания метамодели на валидационной выборке
    meta_preds = meta_model.predict_proba(X_valid_meta)[:, 1]

    # Расчет метрики AUC
    auc = roc_auc_score(y_valid_meta, meta_preds)
    return auc
```

```python
import time

# Сохранение времени начала для первой итерации
start_time = time.time()

def print_progress(study, trial):
    global start_time
    # Проверка, является ли номер испытания кратным 5
    if trial.number % 10 == 0:
        # Вычисление времени, затраченного с момента предыдущей 5-й итерации (или начала)
        elapsed_time = time.time() - start_time
        start_time = time.time()  # Обновляем стартовое время для следующей итерации

        print(f"Trial {trial.number} finished with value: {trial.value} and parameters: {trial.params}")
        print(f"Time elapsed for last 10 trials: {elapsed_time:.2f} seconds")
        print(f"Best trial so far: Trial {study.best_trial.number} with value: {study.best_trial.value}")

optuna.logging.set_verbosity(optuna.logging.WARNING)
```

```python
# Оптимизация гиперпараметров метамодели
study = optuna.create_study(direction="maximize")
study.optimize(objective, n_trials=50, callbacks=[print_progress],n_jobs=-1)
```

```python
final_meta_model = LogisticRegression(**{'C': 0.0007088981995768275, 'solver': 'lbfgs'}, random_state=42, max_iter=1000)
final_meta_model.fit(meta_features, y_valid_meta)
```

```python
# Оценка метрики на тестовой выборке
final_preds = final_meta_model.predict_proba(meta_features)[:, 1]
final_auc = roc_auc_score(y_valid_meta, final_preds)
print(f"AUC финальной метамодели: {final_auc}")
```

```python
import pandas as pd

y_pred = [i[1] for i in final_meta_model.predict_proba(meta_features)]

result_df = pd.DataFrame({
    "id": df_hack_test["id"],
    "target": y_pred
})

result_df.to_csv('submission.csv', index=False)
```
