### 3. Model

#### GPU CatBoost

```python
from catboost import CatBoostClassifier
import optuna
from sklearn.metrics import roc_auc_score
import json
import os
import numpy as np

def objective(trial):
    params = {
        "iterations": trial.suggest_int("iterations", 500, 3000, step=500),  # диапазон итераций
        "learning_rate": trial.suggest_float("learning_rate", 1e-3, 0.1, log=True),  # скорость обучения
        "depth": trial.suggest_int("depth", 4, 12),  # глубина дерева
        "min_data_in_leaf": trial.suggest_int("min_data_in_leaf", 1, 100),  # минимальное количество данных в листе
        "bootstrap_type": trial.suggest_categorical("bootstrap_type", ["Bayesian", "Bernoulli"]),  # тип бутстрепа
        "grow_policy": trial.suggest_categorical("grow_policy", ["Depthwise", "SymmetricTree", "Lossguide"]),  # стратегия роста деревьев
        "use_best_model": trial.suggest_categorical("use_best_model", [True]),  # использование лучшей модели
        "od_wait": trial.suggest_int("od_wait", 10, 50, step=10),
        "l2_leaf_reg": trial.suggest_float("l2_leaf_reg", 1e-1, 10, log=True),  # регуляризация L2
        "scale_pos_weight": trial.suggest_float("scale_pos_weight", 1, 20),  # вес положительного класса для несбалансированных классов
        "max_bin": trial.suggest_int("max_bin", 32, 512),  # максимальное количество бинов для признаков
        "thread_count": trial.suggest_int("thread_count", 1, 16),  # количество потоков для обучения
        "random_strength": trial.suggest_float("random_strength", 0.1, 10, log=True),  # случайность для узлов дерева
        "task_type": trial.suggest_categorical("task_type", ["GPU"]),  # тип устройства (GPU для ускорения обучения)
        "devices": trial.suggest_categorical("devices", ["0"]),  # использование устройства GPU (если оно доступно),
        "eval_metric": trial.suggest_categorical("eval_metric", ["Logloss"]), # метрика для оптимизации 
    }

    if params["bootstrap_type"] != "Bayesian":
        params["subsample"] = trial.suggest_float("subsample", 0.05, 1.0)

    model = CatBoostClassifier(**params, silent=True)
    model.fit(X_train, y_train, eval_set=[(X_val_2, y_val_2)])
    predictions = [i[1] for i in model.predict_proba(X_val_3)]
    roc_auc = roc_auc_score(y_val_3, predictions)
    return roc_auc

if not os.path.exists('models/gpu'):
    os.makedirs('models/gpu')

if not os.path.exists('metrics/gpu'):
    os.makedirs('metrics/gpu')

count = 0
seed = 0 

while count < 20:
    np.random.seed(seed)
    study_gpu = optuna.create_study(direction='maximize')
    study_gpu.optimize(objective, n_trials=100)

    params = study_gpu.best_params
    params['random_state'] = seed

    model_gpu = CatBoostClassifier(**params)
    model_gpu.fit(X_train, y_train, eval_set=(X_val_2, y_val_2), silent=True)

    y_pred = [i[1] for i in model_gpu.predict_proba(X_test)]
    roc_auc = roc_auc_score(y_test, y_pred)
    print(roc_auc)
    if roc_auc > 0.81:
        with open(f'metrics/gpu/roc_auc_score_{count}.txt', 'w') as fp:
            fp.write(str(roc_auc))
        model_gpu.save_model(f'models/gpu/best_model_gpu_{count}', format="cbm")
        count += 1
    seed += 1
```
